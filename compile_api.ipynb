```python
import json

# Define file paths for both JSON files
sherdog_file_path = r'C:\Users\tifek\GitHub\UFC\sherdog.json'
ufc_file_path = r'C:\Users\tifek\GitHub\UFC\ufc.json'

# Load the contents of 'sherdog.json'
with open(sherdog_file_path, 'r') as sherdog_file:
    sherdog_data = json.load(sherdog_file)

# Load the contents of 'ufc.json'
with open(ufc_file_path, 'r') as ufc_file:
    ufc_data = json.load(ufc_file)

# Create a dictionary for easy look-up by name for both datasets
sherdog_dict = {fighter['name']: fighter for fighter in sherdog_data}
ufc_dict = {fighter['name']: fighter for fighter in ufc_data}

# Perform a left join: For each fighter in sherdog, add the corresponding UFC data if available
joined_data = []
for fighter in sherdog_data:
    name = fighter['name']
    ufc_fighter = ufc_dict.get(name)  # Look for the fighter in the ufc_data dictionary
    if ufc_fighter:
        # Merge data from both sources and adjust the keys
        merged_fighter = {
            "name": fighter["name"],
            "sherdog_url": fighter["url"],  # Rename 'url' to 'sherdog_url'
            "ufc_url": ufc_fighter["profile_url"]  # Rename 'profile_url' to 'ufc_url'
        }
    else:
        # If there's no match, just keep the sherdog data
        merged_fighter = {
            "name": fighter["name"],
            "sherdog_url": fighter["url"],  # Rename 'url' to 'sherdog_url'
            "ufc_url": None  # No matching UFC data
        }
    joined_data.append(merged_fighter)

# Print the result as formatted JSON (first 20 entries for readability)
print("Joined Data (First 20 entries):")
print(json.dumps(joined_data[:20], indent=2))
```
